Demonstrating how to combined multiple shapefiles into a single shapefile (using `pandas.concat`)

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt  #not needed for combine, just visual checks
```

```python
#read shapefilesfiles
GIA1 = gpd.read_file("data-working/WoE/GI_Area_1_Priorities/GI_Area_1_Priorities.shp")
GIA16 = gpd.read_file("data-working/WoE/GI_Area_16_Priorities/GI_Area_16_Priorities.shp")
GIA19 = gpd.read_file("data-working/WoE/GI_Area_19_Priorities/GI_Area_19_Priorities.shp")
```

```python
#check files have the same crs
print(GIA1.crs == GIA16.crs)
print(GIA1.crs == GIA19.crs) 
```

```python
#visualise what combined shapefile should eventually look like
fig, ax1 = plt.subplots(1, 1)

GIA1.plot(edgecolor="red", facecolor="None", ax=ax1)
GIA16.plot(edgecolor="blue", facecolor="None", ax=ax1)
GIA19.plot(edgecolor="green", facecolor="None", ax=ax1)
```

```python
#combine
gdf = gpd.GeoDataFrame(pd.concat([GIA1, GIA16, GIA19]))
```

```python
#check visually
gdf.plot(edgecolor="red", facecolor="None")
```

```python
#write to file
gdf.to_file("data-working/WoE/Combined/CombinedAreas-1-16-19.shp")
```

```python
#check file created
comb = gpd.read_file("data-working/WoE/Combined/CombinedAreas-1-16-19.shp")
comb.plot(edgecolor="blue", facecolor="None")
```

```python
#all areas combined
allAreas = gpd.read_file("data-working/WoE/Combined/WOE_GIAreas - Copy.shp")
allAreas.plot(edgecolor="green", facecolor="None")
```
